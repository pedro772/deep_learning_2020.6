```
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/My Drive/Colab Notebooks/shark.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

hist1 = cv2.calcHist([gray], [0], None, [128], [0,128])

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()

sub = gray[210, 600]

for i in range(190, 270):
  for j in range(480, 620):
    gray[i, j] = sub

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()

hist2 = cv2.calcHist([gray], [0], None, [128], [0,128])

fig = plt.figure(facecolor='white')
plt.plot(hist1)
plt.title('Histograma da imagem com caiaque')
plt.show()
plt.savefig("/content/drive/My Drive/Colab Notebooks/histograma_com_caiaque.svg")

fig = plt.figure(facecolor='white')
plt.plot(hist2)
plt.title('Histograma da imagem sem caiaque')
plt.show()
plt.savefig("/content/drive/My Drive/Colab Notebooks/histograma_sem_caiaque.svg")

fig = plt.figure(facecolor='white')
plt.plot(hist1), plt.plot(hist2)
plt.title('Comparação dos 2 histogramas')
plt.show()
plt.savefig("/content/drive/My Drive/Colab Notebooks/comparacao_hists.svg")
```
